**Preprocessing Snort Logs part-2**

```python
import pandas as pd
import datetime as dt
from datetime import datetime
```

```python
tcplogDF = pd.read_csv('tcplog.csv', names = ['datetime','duration','protocol','src','src_port','dst','dst_port','src_bytes','dst_bytes','Unnamed'])
tcplogDF = tcplogDF[tcplogDF.index > 0]
tcplogDF
```

```python
tcplogDF = tcplogDF[['datetime','duration','protocol','src','src_port','dst','dst_port','src_bytes','dst_bytes']]
tcplogDF['datetime_format'] = pd.to_datetime(tcplogDF['datetime'],format="%m/%d-%H:%M:%S.%f")
tcplogDF
```

```python
dstList = tcplogDF['dst'].to_list()
i = 0
count = []
for dtime in tcplogDF['datetime_format'].to_list():
    count.append(len(tcplogDF.loc[(tcplogDF['datetime_format'] <= dtime) & (tcplogDF['datetime_format'] > (dtime - dt.timedelta(seconds=2))) & (tcplogDF['dst']==dstList[i])]))
    i += 1
tcplogDF['count'] = count
tcplogDF
```

```python
dst_portList = tcplogDF['dst_port'].to_list()
i = 0
dst_srv_count = []
for dtime in tcplogDF['datetime_format'].to_list():
    dst_srv_count.append(len(tcplogDF.loc[(tcplogDF['datetime_format'] <= dtime) & (tcplogDF['datetime_format'] > (dtime - dt.timedelta(seconds=2))) & (tcplogDF['dst_port']==dst_portList[i])]))
    i += 1
tcplogDF['srv_count'] = dst_srv_count
tcplogDF
```

```python
tcplogDF = tcplogDF[['datetime','duration','protocol','src','src_port','dst','dst_port','src_bytes','dst_bytes','count','srv_count']]
tcplogDF
```
